```python
from flightdata import Flight
from flightanalysis import Section, FlightLine
from flightanalysis.flightline import Box
from geometry import GPSPosition
import plotly.graph_objs as go
from flightplotting.plots import create_3d_plot, trace
```

```python
fl_definition_log = Flight.from_csv("/home/tom/Desktop/logs/00000002.csv")

pilot = fl_definition_log.subset(0, 23).data.mean()
centre = fl_definition_log.subset(35, 55).data.mean()
pilot = GPSPosition(pilot.global_position_latitude, pilot.global_position_longitude)
centre = GPSPosition(centre.global_position_latitude, centre.global_position_longitude)

flightline = FlightLine.from_box(box.from_points("2point", pilot, centre))
```

```python
sec = Section.from_flight(fl_definition_log, flightline)
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=sec.pos.x, y=sec.pos.y))
fig.update_yaxes(
    scaleanchor = "x",
    scaleratio = 1,
  )
fig
```
